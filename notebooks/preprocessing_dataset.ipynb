# 🛠️ DATA PREPROCESSING

```python
# Separate features and target
TARGET = 'Genre'

# Make a copy, so we don't need to change the orginal DataFrame
df_movie = df.copy()

# Encode target separately so we can decode later
le_target = LabelEncoder()
df_movie[TARGET] = le_target.fit_transform(df_movie[TARGET])
print(f"\nTarget classes: {list(le_target.classes_)}")
print("✅ Encoding done!")
```

# ✂️ SPLIT THE DATASET INTO: X_train and X_test

```python

import sys
import os
from sklearn.model_selection import train_test_split
sys.path.append(os.path.abspath(".."))
from src.preprocessing_dataset import DataPreprocessor




X = df_movie.drop("Genre", axis=1)
y = df_movie["Genre"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=9
)

preprpcessor = DataPreprocessor()
preprpcessor.find_columns(X_train)

X_train_enc, X_test_enc = preprpcessor.encode(
    X_train,
    X_test
)
X_train_scaled, X_test_scaled = preprpcessor.scale(
    X_train_enc,
    X_test_enc
)
```
