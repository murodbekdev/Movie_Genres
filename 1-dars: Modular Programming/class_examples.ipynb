# Bunda BankAccount degan class yasash va unda hozirgi balans, pul qo'shi, pul ayrish va ohiridagi total balans ko'rsatadigan dastur yaratish.

```python
class BankAccount: 
    def __init__(self, balance):
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def show_balance(self):
        print(f"Current Amount: {self.balance}")
    
account = BankAccount(100000)
account.show_balance()
account.deposit(10000)
account.show_balance()
account.withdraw(50000)
account.show_balance()
```

# StudentGrades class yarating va unda talabalarning ismi, baxolar royxati bolsin va shunga kora ularning Ichida eng past, eng yuqori olgan talaba va o’rtacha ballni qaytarsin


```python
class StudentGrades:
    def __init__(self,students):
        self.students = students 
        
    def highest_grade(self):
        print(max(self.students.values()))
    def lowest_grade(self):
        print(min(self.students.values()))
    def average_score(self):
        print(sum(self.students.values())/len(self.students))
students = {
    "Annom": 70,
    "Garab":90,
    "Sherbek":95,
    "Bek":50
}
student = StudentGrades(students)
student.highest_grade()
student.lowest_grade()
student.average_score()
```

```python
class AboutMe:
    def __init__(self, name, age, course, thesis_paper):
        self.name = name 
        self.age = age
        self.course = course
        self.thesis_paper = thesis_paper
    def introduce_myself(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Course: {self.course}")
        print(f"Thesis Paper: {self.thesis_paper}")
about_me = AboutMe("Murodjon", 31, "ML/DL Engineering", "Heart Disease")
about_me.introduce_myself()
```
